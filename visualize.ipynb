```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
with open('download/data.json') as f:
    raw = json.load(f) #['Data']
```

```python
pd.DataFrame(raw)
```

```python
df = pd.DataFrame(raw['Data'])
df.drop(['Detail'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Number of people whom we don't know their age
df['Age'].isnull().sum() / df.shape[0] * 100
```

```python
# Number of reported cases each month
plt.figure(figsize = (10,5))
df['ConfirmDate'] = df['ConfirmDate'].astype('datetime64')
df['ConfirmDate'].groupby([df['ConfirmDate'].dt.year, df['ConfirmDate'].dt.month]).count().plot()
plt.xlabel('Time (Year, Month)')
plt.ylabel('Number of Cases')
plt.title('Number of Reported Cases Each Month')
plt.show()
```
